```python
# Case 1: ANN for Career Recommendation
import tensorflow as tf
import numpy as np

# Input values (normalized)
x1 = 0.8  # Mathematics score
x2 = 0.6  # Communication skill

# Hidden layer parameters
w1 = 0.6
w2 = 0.4
b1 = -0.2

# Output layer weights
w_engineering = 0.9
w_management = 0.6
w_arts = 0.3

# Step 1: Net input to hidden neuron
net_hidden = (x1 * w1) + (x2 * w2) + b1
print("Net input to hidden neuron:", net_hidden)

# Step 2: ReLU activation
hidden_output = tf.nn.relu(net_hidden).numpy()
print("Hidden layer output after ReLU:", hidden_output)

# Step 3: Raw output scores
raw_outputs = np.array([
    hidden_output * w_engineering,
    hidden_output * w_management,
    hidden_output * w_arts
])
print("Raw output scores:", raw_outputs)

# Step 4: Softmax activation
probabilities = tf.nn.softmax(raw_outputs).numpy()
print("\nCareer Probabilities:")
print("Engineering:", probabilities[0])
print("Management :", probabilities[1])
print("Arts       :", probabilities[2])

# Step 5: Final prediction
careers = ["Engineering", "Management", "Arts"]
recommended_career = careers[np.argmax(probabilities)]
print("\nRecommended Career:", recommended_career)
```

```python
# Case 2: ANN for Online Purchase Prediction
import tensorflow as tf
import numpy as np

# Input values (normalized)
x1 = 0.6  # Discount offered
x2 = 0.7  # Customer interest level

# Hidden layer parameters
w1 = 0.5
w2 = 0.7
b1 = -0.3

# Output layer parameters
w_out = 0.8
b2 = -0.1

# Step 1: Net input to hidden neuron
net_hidden = (x1 * w1) + (x2 * w2) + b1
print("Net input to hidden neuron:", net_hidden)

# Step 2: ReLU activation
hidden_output = tf.nn.relu(net_hidden).numpy()
print("Hidden layer output after ReLU:", hidden_output)

# Step 3: Net input to output neuron
net_output = (hidden_output * w_out) + b2
print("Net input to output neuron:", net_output)

# Step 4: Sigmoid activation
purchase_probability = tf.nn.sigmoid(net_output).numpy()
print("\nPurchase Probability:", purchase_probability)

# Step 5: Interpretation
if purchase_probability >= 0.5:
    print("Prediction: Customer is likely to PURCHASE the product")
else:
    print("Prediction: Customer is NOT likely to purchase the product")
```
